```python
# open the record.json file and load the existing data

import json

file = open('records.json', 'r')
data = file.read()
record = json.loads(data)
file.close()

while(True):

    print("\n*****************************")
    print("1. Employee\n2. Customer\n3. Exit")
    print("*****************************")
    ch = int(input("\nPlease, enter your choice: "))
    
    # choice 1. Employee
    if ch == 1:
        print("\n\n************Manage Product List************")
        print("1. Insert new products to the inventory.")
        print("2. Delete the products from the inventory.")
        print("*******************************************")
    
        choice = int(input("Enter your choice: "))
    
        # add new products to the list
        if choice == 1:
            num = int(input("How many entries do you want to make ?"))

            for i in range(0, num):
    
                product_id = input("Enter product id: ")
                name = input("Enter product name: ")
                cost = int(input("Enter the cost: "))
                quantity = int(input("Enter the quantity: "))
                print("\n")
    
                if product_id in record:
                    record[product_id]['qn'] = record[product_id]['qn'] + quantity
                    record[product_id] = {'name':name, 'pr': cost, 'qn': record[product_id]['qn']}
                    print('Product already exists!\nQuantity updated!')
                else:
                    record[product_id] = {'name':name, 'pr': cost, 'qn': quantity}
    
            js = json.dumps(record)
            file = open('records.json', 'w')
            data = file.write(js)
            file.close()
            
            continue
            
        # delete existing products from the list
        elif choice == 2:
            file = open('records.json', 'r')
            data = file.read()
            record = json.loads(data)
            file.close()

            delete = int(input('Enter the number of records you want to delete ?'))

            for i in range(0, delete):
                product_id = input("Enter the product id of the product you want to delete: ")
                if product_id in record:
                    del record[product_id] 
                    print('Deleted successfully!')
                else:
                    print('Product with Id', product_id,'not found!')
                    break

            js = json.dumps(record)
            file = open('records.json', 'w')
            file.write(js)
            file.close()
            
            continue

            
    # choice 2. Customer
    elif ch == 2:

        # printing the product list
        print("*********************** Welcome Customer ****************************")
        print("List of available products: \n")
        for prod in record:
            print(int(prod))
            print('Product: ',record[prod]['name'], '\nCost: ',record[prod]['pr'], '\n')

        
        # purchasing product
        fd = open('sales.json', 'r')
        data = fd.read()
        rec = json.loads(data)
        fd.close()

        import time
        t = time.ctime()

        Total = 0 
        count = len(rec)+1
    
        product_id = input('Enter the product Id: ')
        cust_quantity = int(input('Enter the quantity you want to purchase: '))
        print("\n\n")

        for prod in range(len(record)):
            if product_id not in record:
                print('Product not found!')
                break

            else:
                Total = record[product_id]['pr'] * cust_quantity
        
                print("************ Bill *************")
                print('Bill No.: ', count)
                print('Name: ', record[product_id]['name'])
                print('Cost: ', record[product_id]['pr'])
                print('Quantity: ',cust_quantity)
                print("*******************************")
                print('Total Cost: ', Total)
                print(t)
                break
            
        record[product_id]['qn'] = record[product_id]['qn'] - cust_quantity
        js = json.dumps(record)
        file = open('records.json', 'w')
        file.write(js)
        file.close()
    
        rec[str(count)] = {"prod_id": product_id, "name": record[product_id]['name'], "pr": record[product_id]['pr'], "cust_qn": cust_quantity, "total_cost": Total, "time":t}

        js1 = json.dumps(rec)
        fd = open('sales.json', 'w')
        fd.write(js1)
        fd.close()
        
        continue
        
        
    elif ch == 3:
        print("Thank you!")
        break
        
    else: 
        exit()

        
```
